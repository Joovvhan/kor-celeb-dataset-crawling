```python
# -*- coding: utf-8 -*-
import os
import shutil
import glob
from tqdm.notebook import tqdm
```

```python
dataset_path = 'D:\Korean Celeb Data'

# with open('../kor_female_idol_list.txt', encoding='UTF-8') as f:
#     lines = f.readlines()
#     keywords = [line.strip() for line in lines]

keywords = os.listdir(dataset_path)
```

```python
for keyword in tqdm(keywords):
    celeb_dir = os.path.join(dataset_path, keyword)
    
    if os.path.isdir(celeb_dir):
        dirs = [file_name for file_name in os.listdir(celeb_dir) if os.path.isdir(os.path.join(celeb_dir, file_name))]
        crawled_img_dirs = [os.path.join(celeb_dir, direc) for direc in dirs]
        
        for direc in crawled_img_dirs:
            prefix = direc.split('\\')[-1][0:2]
            files = os.listdir(direc)
            
            for file in files:
                src = os.path.join(direc, file)
                dst = os.path.join(celeb_dir, prefix + '_' + file)
                shutil.move(src, dst)
                
            os.rmdir(direc)
            print('Removed {}'.format(direc))
```
